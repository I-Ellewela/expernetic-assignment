```python
# ============================================
# SECTION 02 - DATASET FAMILIARIZATION
# Expernetic Data Engineering Assignment
# Candidate: Ishini Ellewela
# Dataset: Inside Airbnb - Bangkok, Thailand
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Paths
data_path = r"C:\Users\Ishini\Desktop\Expernetic_Assignment\data"
charts_path = r"C:\Users\Ishini\Desktop\Expernetic_Assignment\charts"

print("All libraries imported successfully!")
print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")
```

```python
# List all files in the data folder
files = os.listdir(data_path)
print("Files in dataset folder:")
for f in files:
    size = os.path.getsize(os.path.join(data_path, f))
    print(f"  {f} — {size/1024:.1f} KB")
```

```python
# Load the listings dataset
listings = pd.read_csv(os.path.join(data_path, "listings.csv.gz"), compression='gzip')

print("=== LISTINGS DATASET OVERVIEW ===")
print(f"Rows: {listings.shape[0]:,}")
print(f"Columns: {listings.shape[1]}")
print(f"\nColumn Names:")
for col in listings.columns:
    print(f"  - {col}")
```

```python
# Document schema - data types, missing values, sample values
print("=== SCHEMA DOCUMENTATION ===\n")
print(f"{'Column':<45} {'Data Type':<15} {'Non-Null':<10} {'Missing %':<10} {'Sample Value'}")
print("-" * 110)

for col in listings.columns:
    dtype = str(listings[col].dtype)
    non_null = listings[col].count()
    missing_pct = (listings[col].isnull().sum() / len(listings)) * 100
    sample = str(listings[col].dropna().iloc[0])[:30] if listings[col].count() > 0 else "N/A"
    print(f"{col:<45} {dtype:<15} {non_null:<10} {missing_pct:<9.1f}% {sample}")
```

```python
# Load remaining datasets
print("Loading other datasets...\n")

# Neighbourhoods
neighbourhoods = pd.read_csv(os.path.join(data_path, "neighbourhoods.csv"))
print("=== NEIGHBOURHOODS DATASET ===")
print(f"Rows: {neighbourhoods.shape[0]}, Columns: {neighbourhoods.shape[1]}")
print(f"Columns: {list(neighbourhoods.columns)}")
print(neighbourhoods.head())

# Reviews
print("\n=== REVIEWS DATASET ===")
reviews = pd.read_csv(os.path.join(data_path, "reviews.csv.gz"), compression='gzip')
print(f"Rows: {reviews.shape[0]:,}, Columns: {reviews.shape[1]}")
print(f"Columns: {list(reviews.columns)}")
print(reviews.head(3))

# Calendar
print("\n=== CALENDAR DATASET ===")
calendar = pd.read_csv(os.path.join(data_path, "calendar.csv.gz"), compression='gzip')
print(f"Rows: {calendar.shape[0]:,}, Columns: {calendar.shape[1]}")
print(f"Columns: {list(calendar.columns)}")
print(calendar.head(3))
```

```python
# Dataset summary and relationships
print("=== DATASET SUMMARY ===\n")
print(f"Listings:       {listings.shape[0]:>10,} rows × {listings.shape[1]} columns")
print(f"Reviews:        {reviews.shape[0]:>10,} rows × {reviews.shape[1]} columns")
print(f"Calendar:       {calendar.shape[0]:>10,} rows × {calendar.shape[1]} columns")
print(f"Neighbourhoods: {neighbourhoods.shape[0]:>10,} rows × {neighbourhoods.shape[1]} columns")

print("\n=== FILE RELATIONSHIPS ===")
print("listings.id → reviews.listing_id (one listing : many reviews)")
print("listings.id → calendar.listing_id (one listing : 365 calendar days)")
print("listings.neighbourhood_cleansed → neighbourhoods.neighbourhood")

print("\n=== KEY OBSERVATIONS ===")
print(f"1. Bangkok has {neighbourhoods.shape[0]} official districts (khet)")
print(f"2. Average reviews per listing: {reviews.shape[0]/listings.shape[0]:.1f}")
print(f"3. Calendar covers ~{calendar.shape[0]/listings.shape[0]:.0f} days per listing")
print(f"4. Listings with NO reviews: {listings['number_of_reviews'].eq(0).sum():,} ({listings['number_of_reviews'].eq(0).mean()*100:.1f}%)")
print(f"5. Columns with 100% missing values: {(listings.isnull().mean()==1).sum()}")
print(f"6. Price column stored as string - needs cleaning")
print(f"7. neighbourhood_group is 100% empty in neighbourhoods file")

print("\n=== DATASET LIMITATIONS ===")
print("- neighbourhood_overview: 100% missing")
print("- host_response_time, host_response_rate, host_acceptance_rate: 100% missing")
print("- license, instant_bookable, calendar_updated: 100% missing")
print("- review scores: 33% missing (listings with no reviews)")
print("- host_location: 25.8% missing")
print("- price: 6.7% missing (listings without active pricing)")
```
